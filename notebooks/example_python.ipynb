# NextGenDB - Python Quickstart

This notebook demonstrates how to connect to the NextGenDB REST API using Python and perform basic graph and agentic queries.

```python
import requests
import json

API_URL = "http://localhost:8000/api/v1"
```

### 1. Perform a Query

The core endpoint is `/query`. It processes natural language, Cypher, and SQL.

```python
def run_query(query_string):
    response = requests.post(f"{API_URL}/query", json={"query": query_string})
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code}")
        return None

# Cypher query
res = run_query("MATCH (n:USER) RETURN n")
print("Answer:", res.get('answer'))
print("Confidence:", res.get('confidence'))
```

### 2. Inspecting the Graph
The API returns the graph snapshot involved in the query response.

```python
graph = res.get('graph_snapshot', {})
print(f"Nodes: {len(graph.get('nodes', []))}")
print(f"Edges: {len(graph.get('links', []))}")
```

### 3. Checking Observability Metrics
Retrieve Prometheus-compatible JSON metrics from the dashboard.

```python
metrics_res = requests.get(f"{API_URL}/metrics/dashboard")
if metrics_res.status_code == 200:
    data = metrics_res.json()
    print("\n--- System Metrics ---")
    print(json.dumps(data['metrics'], indent=2))
    print("\n--- Layer Status ---")
    for layer, status in data['layer_status'].items():
        print(f"{layer}: {status}")
```
